```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd #to clean
```

```python
# Load the Excel file
excel_file = pd.ExcelFile('datos_actividad2.xlsx')

# Parse each sheet into a DataFrame
df = excel_file.parse('Hoja2')
df2 = excel_file.parse('Hoja3')
```

```python
#See the variables of the first sheet
print(list(df))
```

```python
# Columns to drop by index
columns_to_drop = list(range(13)) + [14, 15, 26]
```

```python
drop_df=df.drop(df.columns[columns_to_drop], axis=1)
```

```python
drop_df.head(2)
```

```python
#Calculate age based on 'LastUpdated' and 'BirthDate'

# Convert 'BirthDate' and 'LastUpdated' columns to datetime
drop_df['BirthDate'] = pd.to_datetime(drop_df['BirthDate'])
drop_df['LastUpdated'] = pd.to_datetime(drop_df['LastUpdated'])
```

```python
# Calculate age by subtracting the year of 'BirthDate' from 'LastUpdated'
drop_df['Age'] = drop_df['LastUpdated'].dt.year - drop_df['BirthDate'].dt.year
drop_df.head(2)
```

```python
# List of column indices to drop
columns_to_drop = [0, 8]
```

```python
# Drop the specified columns
drop_df = drop_df.drop(drop_df.columns[columns_to_drop], axis=1)
drop_df.head(2)
```

```python
#we are going to encode the 'Gender' and 'MaritalStatus'
encoded_df = pd.DataFrame(drop_df)

# Defining mapping dictionary for 'Gender' and 'MaritalStatus'
gender_mapping = {'M': 0, 'F': 1}
marital_mapping = {'S': 0, 'M': 1}
```

```python
#replace to map the values
encoded_df['Gender'] = encoded_df['Gender'].replace(gender_mapping)
encoded_df['MaritalStatus'] = encoded_df['MaritalStatus'].replace(marital_mapping)
```

```python
encoded_df.head(2)
```

```python
list(encoded_df)
```

```python
# Move 'BikeBuyer' column to the last position

cols = list(encoded_df.columns)
if 'BikeBuyer' in cols:
    cols.remove('BikeBuyer')
cols = cols + ['BikeBuyer']  # Move 'BikeBuyer' to last position
encoded_df = encoded_df[cols]
```
